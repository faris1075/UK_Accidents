# INTRODUCTION TO EXPLORATORY DATA ANALYSIS

Exploratory Data Analysis, often abbreviated as EDA, is a crucial first step in any data analysis project. It's like being a detective investigating a new case - you examine the evidence (data) carefully, trying to understand its main characteristics, identify patterns, and uncover interesting relationships.

#### BOXPLOT OF POLICE FOURCE COLUMN

```python
df[['Police_Force']].boxplot()
plt.title('POLICE FOURCE')
plt.show()
```

### BAR PLOTTING IS YEAR-BASED ACCIDENTS INDICATING

```python
df.Year.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
font1={'size':20}
font2={'size':15}
plt.title("UK ACCIDENT",fontdict=font1)
plt.ylabel('Number of Accident',fontdict=font2)
plt.xlabel('Year',fontdict=font2)
plt.show()
```

```python
df1=df[['Year','Speed_limit']]
df1
```

### LINE PLOT YEAR-BASED SPEED LIMIT INDICATING

```python
df1=df[['Year','Speed_limit']]
sns.lineplot(x='Year',y='Speed_limit',data=df1,color='green')
sns.despine()
plt.title("SPEED LIMIT")
plt.show()
```

### SCATTER PLOT NUMBER OF CASUALTIES TO ROAD CONDITIONS

```python
plt.scatter(df['Number_of_Casualties'],df['Road_Conditions'], color='b')
plt.colorbar()
plt.title("Casualties")
plt.show()
```

### PIE CHART

### HISTPLOT

```python
Police_attend = df['Police_Officer_Attend_Scene_of_Accident'].mean() * 100
print(f"Situation : {Police_attend:.2f}%")

df[df['Police_Officer_Attend_Scene_of_Accident'] == True]['Urban_or_Rural_Area'].hist(alpha=0.5, label=' Attented',color='g')
df[df['Police_Officer_Attend_Scene_of_Accident'] == False]['Urban_or_Rural_Area'].hist(alpha=0.5, label='Not Attented')
plt.legend()
plt.show()
```
